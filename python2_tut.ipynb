```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("/home/Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df=df.drop(columns="Not_Useful_Column")
df
```

```python
df["Last_Name"]=df["Last_Name"].str.lstrip("...")
df["Last_Name"]=df["Last_Name"].str.lstrip("/")
df["Last_Name"]=df["Last_Name"].str.rstrip("_")
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]',' ',regex=True)
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace(' ','-',regex=True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].fillna('')
df["Phone_Number"] = df["Phone_Number"].str.replace('N-a', '', regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace('NaN', '', regex=True)
df
```

```python
df[["street","state", "zip_code"]]=df["Address"].str.split(',', n=2, expand=True)
df
```

```python
df=df.drop(columns="Address")
df
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("N", "No")
df["Paying Customer"]=df["Paying Customer"].replace("Y", "Yes")
df["Paying Customer"]=df["Paying Customer"].replace("N/a", "")
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("Y","Yes")
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("N","No")
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace(np.NaN,"")
df
```

```python
df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Yes':
         df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
         df.drop(x, inplace=True)

df
```
