```python
import pandas as pd
df1=pd.read_csv("/home/LOTR.csv")
df1
```

```python
df2=pd.read_csv("/home/LOTR 2.csv")
df2
```

```python
df1.merge(df2, how = 'inner', on=["FellowshipID","FirstName"])
```

```python
df1.merge(df2, how = 'outer', on=["FellowshipID","FirstName"])
```

```python
df1.merge(df2, how = 'left', on=["FellowshipID","FirstName"])
```

```python
df1.merge(df2, how = 'right', on=["FellowshipID","FirstName"])
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2,  on="FellowshipID", how="outer", lsuffix='_left', rsuffix="_right")
```

```python
df3=df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_viv', rsuffix='_bhavi', how="inner")
```
